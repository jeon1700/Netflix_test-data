```python
# 장르 분할 4시간째 안된다...(
```

```python
df['genres'] = df['genres'].apply(lambda x: x.strip("[]").replace("'", "").split(", ") if pd.notna(x) else [])
```

```python
df['genres'] = df['genres'].apply(lambda x: ['Normal'] if all(genre.strip() == '' for genre in x) else x)
```

```python
all_genres = [genre for sublist in df['genres'] for genre in sublist]
```

```python
genre_counts = pd.Series(all_genres).value_counts()
```

```python
genre_counts.loc['Normal'] = 0
```

```python
for genre in genre_counts.index:
    df[f'genre_{genre}'] = df['genres'].apply(lambda x: genre in x).astype(int)
```

```python
df = df.drop('genres', axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = X_train.fillna(0)
```

```python
netfl= LinearRegression()
```

```python
netfl.fit(X_train, y_train)
```

```python
y_pred = netfl.predict(X_test)
```
